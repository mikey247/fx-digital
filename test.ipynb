```python

def round_values(rgb_values):
    r,g,b = rgb_values

    r = round(r/10) * 10
    g = round(g/10) * 10
    b = round(b/10) * 10

    r = min(255, max(0, r))
    g = min(255, max(0, g))
    b = min(255, max(0, b))

    print(r, g, b)


round_values((33, 56, 255))
```

```python
from PIL import Image

# Make a 4x4 image, all red to start. (width, height), then the fill colour.
img = Image.new("RGB", (40, 40), (255, 0, 0))

# Overwrite a few pixels with blue using (x, y) coordinates.
img.putpixel((0, 0), (0, 0, 255))
img.putpixel((1, 0), (0, 0, 255))

# save image
img.save("test.png")
```

```python
img = Image.open('test.png')
img.convert('RGB')

width, height = img.size
print(width, height)
```

```python
for x in range(0, width):
    for y in range(0, height):
        print(img.getpixel((x,y)))
```
